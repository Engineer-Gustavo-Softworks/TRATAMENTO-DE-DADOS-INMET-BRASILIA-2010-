>Removendo outliers da amostra:


def remove_outliers_iqr(df):

    df_cleaned = df_copia.copy()

    for column in df_cleaned.select_dtypes(include=[np.number]).columns:

        # Calcule o Q1 (1 quartil) e Q3 (3 quartil)
        Q1 = df_cleaned[column].quantile(0.25)
        Q3 = df_cleaned[column].quantile(0.75)

        # Calcule o IQR (Intervalo Interquartil)
        IQR = Q3 - Q1

        # Determine os limites inferior e superior para os outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filtre os dados para remover outliers
        df_cleaned = df_cleaned[(df_cleaned[column] >= lower_bound) & (df_cleaned[column] <= upper_bound)]

    return df_cleaned

```python
# Combinar data + hora
df_copia['DATA HORA'] = pd.to_datetime(df_copia['DATA (YYYY-MM-DD)'].astype(str) + ' ' + df_copia['HORA (UTC)'].astype(str))
```

```python
df_copia.head()
```

```python
#Criando uma coluna de 'ANO MES' para agrupar
df_copia['ANO MES'] = df_copia['DATA HORA'].dt.to_period('M')
```

```python
#Agrupando e calculando a mdia
df_copia = df_copia.groupby('ANO MES')['VENTO, VELOCIDADE HORARIA (m/s)'].mean().reset_index()
```
